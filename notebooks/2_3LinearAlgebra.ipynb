```python
matrix = torch.arange(20, dtype = torch.float32).reshape(4, 5)
matrix
```

```python
matrix.T
```

```python
n = 2
n * matrix
```

```python
# axis 表示要消解的维度
# 进行一次求和时，保持其他维度下标不变，对该维度下下标的所有值对应的数求和
print(matrix.sum())
print(matrix.sum(axis = 0))
print(matrix.sum(axis = 1))
print(matrix.sum(axis = [0, 1]))
```

```python
# mean 要求 元素为浮点数或复数
matrix.mean(), matrix.sum() / matrix.numel()
```

```python
# 保持轴不变
sum = matrix.sum(axis = 1, keepdim = True)
sum
```

```python
matrix / sum
```

```python
# 累积总和
matrix.cumsum(axis = 1)
```
